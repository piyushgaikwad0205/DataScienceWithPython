```python
from scipy.stats import pearsonr
```

#Co-Relation and P Value

> If Co-Relation was Greater and P value was Less then it's called it's Co-relation was Significated

> If Co-Relation was Greater and P Value was Greater then it's Called it's Co-relation was unsignifiicated

>> In Below Cell We Convert Co-relation into P Value

```python
corr , p_value = pearsonr(df['engine-size'] , df['price'])
print(corr.round(2))
print(f"The P Value of The Correlation of Engine Size and Price is {p_value:.7f}")
```
